Importing libraries for modelling and classification of data.

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from pandas.plotting import scatter_matrix
from matplotlib import pyplot
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
```

Importing the csv data from file- Shakespeare_data.csv in pandas dataframe

```python
data = pd.read_csv(r'C:\Stuff\KU Study\EECS 731 Intro to Data Science\Assignment-Project-2---To-be-or-not-to-be\data\external\Shakespeare_data.csv')
```

Dropping the rows with NaN values so we don't have to include the values of other features in the dataset

```python
# data['PlayerLinenumber'] = data['PlayerLinenumber'].fillna(0)
# data['ActSceneLine'] = data['ActSceneLine'].fillna(0)
# data["Player"]= data["Player"].fillna(0)
data.dropna(subset = ["Player"], inplace=True)
data.dropna(subset = ["PlayerLinenumber"], inplace=True)
data.dropna(subset = ["ActSceneLine"], inplace=True)
```

Converting the data in ActSceneLine and Player as str type to remove the conflict of having both string and float values

```python
data["ActSceneLine"]= data["ActSceneLine"].astype(str) 
data["Player"]= data["Player"].astype(str)
```

Splitting the Act, Scene and Line values so we can collaborate act and scene together

```python
data[['Act','Scene','Line']] = data.ActSceneLine.str.split(".",expand=True)
```

Dropping the Dataline, Line, ActSceneLine, PlayerLinenumber and PlayerLine because these features won't matter in predicting the player.

```python
new_data = data.drop(["Dataline","Line","ActSceneLine","PlayerLinenumber","PlayerLine"],axis='columns')
```

Converting Act and Scene into string.

```python
new_data['Act'] = new_data['Act'].astype(str)
new_data['Scene']= new_data['Scene'].astype(str)
```

Joining Play, Act and Scene as a single feature because Act and Scene individually won't have a relationship accross the Plays.

```python
new_data['PlayActScene']=new_data['Play'] + "." + new_data['Act'] + "." + new_data['Scene'] 
new_data.head()
```

Dropping individual Play, Act and Scene columns

```python
new_data=new_data.drop(['Play','Act','Scene'], axis='columns')
new_data.head()
```

Using label encoder to convert the text data in to numerical data so we can feed the data in to the models

```python
le=LabelEncoder()
playActScene=le.fit_transform(new_data["PlayActScene"])

#player=le.fit_transform(new_data["Player"])
# playerLineNumber=le.fit_transform(data["PlayerLinenumber"])
# actSceneLine=le.fit_transform(data["ActSceneLine"])
# playerLine=le.fit_transform(data["PlayerLine"])
```

Dropping the columns ["Play","Player"] from original data with text values and adding the respective columns with dummy numerical values in the new_data.

```python
new_data["PlayActScene"]=playActScene
new_data.head()

# new_data["PlayerLine"]=playerLine
# new_data["PlayerLinenumber"]=playerLineNumber
# new_data["ActSceneLine"]=actSceneLine
```

Separating the features(X) as PlayActScene and output(y) as Player. Removing the Dataline column as it just specify the index that is insignificant to the dataset.

```python
X=new_data.drop(columns=['Player'])
y=new_data['Player']
```

Splitting the features and output data in to Validation and training dataset using 10-fold cross validation technique. 

```python
X_training, X_validation, Y_training, Y_validation = train_test_split(X, y, test_size=0.1, random_state=1)
```

Determining the Accuracy of the model Naive Bayes using 10- fold cross validation .

```python
kfold = StratifiedKFold(n_splits=10, random_state=1, shuffle=True)
cv_results = cross_val_score(GaussianNB(), X_training, Y_training, cv=kfold, scoring='accuracy')
print('%s: %f (%f)' % ('NB Accuracy = ', cv_results.mean(), cv_results.std()))
```

Fitting the Naive Bayes model with training data and then making predictions on the validation data.

```python
model=GaussianNB()
model.fit(X_training,Y_training)
predictions=model.predict(X_validation)
```

Printing the overall precision and Classification report having precision, recall, f1-score and support of individual classes.

```python
print('%s: %f' % ('Precision =' , accuracy_score(Y_validation,predictions)))
print(classification_report(Y_validation,predictions))
```

Determining the Accuracy of the model KNeighborsClassifier model using 10- fold cross validation .

```python
cv_results = cross_val_score(KNeighborsClassifier(), X_training, Y_training, cv=kfold, scoring='accuracy')
print('%s: %f (%f)' % ('KNeighbors = ', cv_results.mean(), cv_results.std()))
```

Fitting the KNeighbors model with training data and then making predictions on the validation data.

```python
model = KNeighborsClassifier()
model.fit(X_training,Y_training)
predictions=model.predict(X_validation)
```

Printing the overall precision and classification report having precision, recall, f1-score and support of individual classes.

```python
print('%s: %f' % ('Precision =' , accuracy_score(Y_validation,predictions)))
print(classification_report(Y_validation,predictions))
```

Saving the resulting data to CSV file.

```python
new_data.to_csv(r"C:\Stuff\KU Study\EECS 731 Intro to Data Science\Assignment-Project-2---To-be-or-not-to-be\data\processed\processed_data.csv",index=False) 
```
